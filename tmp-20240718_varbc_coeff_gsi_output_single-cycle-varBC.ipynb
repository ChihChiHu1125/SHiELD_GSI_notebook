```python
# read and analysis for varBC coefficient GSI output file
# initial built: 2024/07/18

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
root='/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/single_cycle_VarBC_experiment/BC_coefficient/'
exps=['bc0','bc0_inf','bc0_noinf','bc1','bc1_inf','bc1_noinf']

#fn='/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/single_cycle_VarBC_experiment/BC_coefficient/BC1/gdas.t18z.abias'
fn=[];
for exp in exps:
    if exp in ('bc0', 'bc1'): # prior coefficients:
        fn.append(f'{root}/{exp}/gdas.t18z.abias')
    else:
        fn.append(f'{root}/analysis/{exp}/gdas.t00z.abias')
```

```python
keys=['sensor_name','sensor_ch','global_offset','sq_temp','temp', \
      'ang_4th', 'ang_3rd', 'ang_2nd', 'ang_1st']

df_exp={};

for e, exp in enumerate(exps):
    bc_dic = {key: [] for key in keys}

    with open(fn[e], 'r') as file:

        for i, line in enumerate(file):

            if np.mod(i,3)==0:
                parts = line.split()
                if parts[-1] != '0':
                    read_line = True
                    bc_dic['sensor_name'].append(parts[1])
                    bc_dic['sensor_ch'].append(parts[2])
                else:
                    read_line = False

            elif np.mod(i,3)==1 and read_line:
                parts = line.split()
                bc_dic['global_offset'].append(parts[0])
                bc_dic['sq_temp'].append(parts[3])
                bc_dic['temp'].append(parts[4])
                bc_dic['ang_4th'].append(parts[8])
                bc_dic['ang_3rd'].append(parts[9])

            elif np.mod(i,3)==2 and read_line:
                parts = line.split()
                bc_dic['ang_2nd'].append(parts[0])
                bc_dic['ang_1st'].append(parts[1])

    # turn the dictionary into panda frame
    df = pd.DataFrame(bc_dic)
    df_exp[exp] = df
```

```python
# a few panda dataframe operations examples
sensor_select='amsua'
ch_select='1'
exp_select='bc0'

df = df_exp[exp_select]
df_tmp = df[ (df['sensor_name'].str.startswith(sensor_select)) & \
             (df['sensor_ch'] == ch_select )]
df_tmp
```
