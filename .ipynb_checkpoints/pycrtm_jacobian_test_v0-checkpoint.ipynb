```python
# plot pyCRTM output jacobian
# initial built: 2024/06/19

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
npzdir = '/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/pyCRTM/crtm-bundle/pycrtm/output_npz/'
fn     = 'atms_jacobian_test.npz'
data   = np.load(npzdir+fn)
```

```python
variable_names = data.files
print(variable_names)
```

```python
P={}; T={}; Q={}; C={}; SF={};
JQ={}; JT={}; JCr={}; JCc={}; JCf={};

for i in range(4):
    P[i]   = data['P'][i]
    T[i]   = data['T'][i]
    Q[i]   = data['Q'][i]
    C[i]   = data['C'][i]
    SF[i]  = data['SF'][i]

    JQ[i]  = data['JQ'][i]
    JT[i]  = data['JT'][i]
    JCr[i] = data['JCr'][i]
    JCc[i] = data['JCc'][i]
    JCf[i] = data['JCf'][i]
```

```python
# surface features:
SF
```

```python
# the general features of the profiles:
plt_P = P[0];

# temperature
plt.plot(T[0], plt_P, label='prof 0')
plt.plot(T[1], plt_P, label='prof 1')
plt.plot(T[2], plt_P, label='prof 2')
plt.plot(T[3], plt_P, label='prof 3')

plt.ylabel('pressure (hPa)')
plt.xlabel('temperature (K)')
plt.title(f'Temperature profile',fontsize=14)
plt.ylim(1020,0)
#plt.xlim(-2, 2)
plt.grid(True)
plt.legend()
plt.show()

# humidity
plt.plot(Q[0], plt_P, label='prof 0')
plt.plot(Q[1], plt_P, label='prof 1')
plt.plot(Q[2], plt_P, label='prof 2')
plt.plot(Q[3], plt_P, label='prof 3')

plt.ylabel('pressure (hPa)')
plt.xlabel('humidity (g/kg)')
plt.title(f'Humidity profile',fontsize=14)
plt.ylim(1020,0)
#plt.xlim(-2, 2)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# cloud concentration:
plt.plot(C[0][:,0,0], plt_P, label='prof 0')
plt.plot(C[1][:,0,0], plt_P, label='prof 1')
plt.plot(C[2][:,0,0], plt_P, label='prof 2')
plt.plot(C[3][:,0,0], plt_P, label='prof 3')

plt.ylabel('pressure (hPa)')
plt.xlabel('cloud concentration (g/kg)')
plt.title(f'Cloud profile',fontsize=14)
plt.ylim(1020,0)
#plt.xlim(-2, 2)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Humidity Jacobian

for i in range(22):
    plt_ch   = i+1; 
    plt_P = P[0]; # the same for all profiles

    fig = plt.figure(figsize=(6,6))
    plt.plot(JQ[0][plt_ch-1,:],plt_P, label='prof 0')
    plt.plot(JQ[1][plt_ch-1,:],plt_P, label='prof 1')
    plt.plot(JQ[2][plt_ch-1,:],plt_P, label='prof 2')
    plt.plot(JQ[3][plt_ch-1,:],plt_P, label='prof 3')

    plt.axvline(x=0,linestyle='--',color=(0.5,0.5,0.5))
    plt.ylabel('pressure (hPa)')
    plt.xlabel('Jacobian Q (K/g/kg)')
    plt.title(f'Humidity Jacobian for ATMS channel {plt_ch}',fontsize=14)
    plt.ylim(1020,0)
    plt.xlim(-2, 2)
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# Temperature Jacobian

for i in range(22):
    plt_ch   = i+1; 
    plt_P = P[0]; # the same for all profiles

    fig = plt.figure(figsize=(6,6))
    plt.plot(JT[0][plt_ch-1,:],plt_P)
    plt.plot(JT[1][plt_ch-1,:],plt_P)
    plt.plot(JT[2][plt_ch-1,:],plt_P)
    plt.plot(JT[3][plt_ch-1,:],plt_P)

    plt.axvline(x=0,linestyle='--',color=(0.5,0.5,0.5))
    plt.ylabel('pressure (hPa)')
    plt.xlabel('Jacobian T (K/K)')
    plt.title(f'Temperature Jacobian for ATMS channel {plt_ch}',fontsize=14)
    plt.ylim(1020,0)
    plt.xlim(-0.2, 0.2)
    plt.grid(True)
    plt.show()
```
