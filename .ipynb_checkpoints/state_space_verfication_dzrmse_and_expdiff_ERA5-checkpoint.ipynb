```python
# Plot the verification against reanalysis data (for IFS only):
# Plot normalized 'drmse' in lat-height coordinate
# initial built: 2024/09/05
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import netCDF4 as nc
```

```python
datahome    = '/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/GribDiag/ush/plotgribnc/data/gribnc'

timerange = '2022061500_2022072000_12'
#dirname   = 'ctrl_amsua_ch5-only_clr_clr-only_clr_clr_ng-clr_and_cld_0.05-clr_and_cld_0.05_ng'
#dirname  = 'ctrl_amsua_ch5-only_clr_clr-clr_and_cld_0.05-clr_and_cld_0.05_sym-clr_and_cld_0.05_model-clr_and_cld_0.05_obs'
dirname = 'ctrl-ctrl_noinf'

exps      = dirname.split('-')
expref    = exps[0]

ncdir_ifs  = os.path.join(datahome,'IFS', timerange,dirname)
```

```python
exps
```

```python
# read netcdfs:

dr_norm={}; dt_norm={}; du_norm={}; dv_norm={};  # normalized error relative to exps[0]
dr={}; dt={}; du={}; dv={};                      # absolute error relative to IFS
s95_r={}; s95_t={}; s95_u={}; s95_v={};

# normalized error and significant levels:
for exp in exps:
    if exp != expref: 
        # normalized relative error
        f = nc.Dataset(f"{ncdir_ifs}/dzmean_rmse_{exp}_f06.nc")
        lat = f.variables['latitude'][:]
        lev = f.variables['level'][:]
        dr_norm[exp] = f.variables['r'][:]
        dt_norm[exp] = f.variables['t'][:]
        du_norm[exp] = f.variables['u'][:]    
        dv_norm[exp] = f.variables['v'][:]    

        d95 = nc.Dataset(f"{ncdir_ifs}/d95_{exp}_f06.nc")
        d95_r = d95.variables['r'][:]
        d95_t = d95.variables['t'][:]
        d95_u = d95.variables['u'][:]    
        d95_v = d95.variables['v'][:]    

        s95_r[exp] = np.abs(d95_r/dr_norm[exp])>1.0
        s95_t[exp] = np.abs(d95_t/dt_norm[exp])>1.0
        s95_u[exp] = np.abs(d95_u/du_norm[exp])>1.0  
        s95_v[exp] = np.abs(d95_v/dv_norm[exp])>1.0  

    # absolute error
    f = nc.Dataset(f"{ncdir_ifs}/zmean_rmse_{exp}_f06.nc")
    dr[exp] = f.variables['r'][:]
    dt[exp] = f.variables['t'][:]
    du[exp] = f.variables['u'][:]    
    dv[exp] = f.variables['v'][:]   
```

```python
# Temperature:

for exptest in exps[1:]:
#exptest   = 'clr_and_cld_0.05'

    lats, levs = np.meshgrid(lat, lev)

    # normalized temperature
    vmin_norm = -0.05
    vmax_norm =  0.05

    # absolute temperature
    vmin_abs = -0.05
    vmax_abs =  0.05

    # Main Plot section:

    # figure size:
    fig = plt.figure(figsize=(14,4))

    # define the subplots
    ax1 = fig.add_subplot(1, 2, 1)  # Add subplot at position 1
    ax2 = fig.add_subplot(1, 2, 2)  # Add subplot at position 2

    # ax1: normalized error
    contour1=ax1.pcolormesh(lats, levs, dt_norm[exptest], cmap='RdBu_r', vmin=vmin_norm, vmax=vmax_norm)
    ax1.invert_yaxis()
    ax1.plot(np.ma.array(lats, mask=s95_t[exptest]), levs, 'k+', ms=2.5)
    ax1.set_xlabel('latitude')
    ax1.set_ylabel('height (hPa)')
    ax1.set_title(f'norm drmse T ({exptest} - {expref}), Ref=IFS')
    ax1.set_ylim(1000,10)
    ax1.grid(True)
    # set colorbar
    cbar = fig.colorbar(contour1,location='bottom',aspect=40)
    cbar.set_label('normalized drmse (%)')

    # ax2: absoulte error
    contour2 = ax2.pcolormesh(lats, levs, dt[exptest] - dt[expref], cmap='RdBu_r', vmin=vmin_abs, vmax=vmax_abs)
    ax2.invert_yaxis()
    ax2.plot(np.ma.array(lats, mask=s95_t[exptest]), levs, 'k+', ms=2.5)
    ax2.set_xlabel('latitude')
    ax2.set_ylabel('height (hPa)')
    ax2.set_title(f'drmse T ({exptest} - {expref}), Ref=IFS')
    ax2.set_ylim(1000,10)
    ax2.grid(True)
    # set colorbar
    cbar = fig.colorbar(contour2,location='bottom',aspect=40)
    cbar.set_label('drmse (K)')
    
    plt.savefig(f"IFS_drmse_T_{exptest}.png")
```

```python
# Relative Humidity:

for exptest in exps[1:]:
#exptest   = 'clr_and_cld_0.05'

    lats, levs = np.meshgrid(lat, lev)

    # normalized relative humidity
    vmin_norm = -0.05
    vmax_norm =  0.05

    # absolute relative humidity
    vmin_abs = -1
    vmax_abs =  1

    # Main Plot section:

    # figure size:
    fig = plt.figure(figsize=(14,4))

    # define the subplots
    ax1 = fig.add_subplot(1, 2, 1)  # Add subplot at position 1
    ax2 = fig.add_subplot(1, 2, 2)  # Add subplot at position 2

    # ax1: normalized error
    contour1=ax1.pcolormesh(lats, levs, dr_norm[exptest], cmap='RdBu_r', vmin=vmin_norm, vmax=vmax_norm)
    ax1.invert_yaxis()
    ax1.plot(np.ma.array(lats, mask=s95_r[exptest]), levs, 'k+', ms=2.5)
    ax1.set_xlabel('latitude')
    ax1.set_ylabel('height (hPa)')
    ax1.set_title(f'norm drmse RH ({exptest} - {expref}), Ref=IFS')
    ax1.set_ylim(1000,10)
    ax1.grid(True)
    # set colorbar
    cbar = fig.colorbar(contour1,location='bottom',aspect=40)
    cbar.set_label('normalized drmse (%)')

    # ax2: absoulte error
    contour2 = ax2.pcolormesh(lats, levs, dr[exptest] - dr[expref], cmap='RdBu_r', vmin=vmin_abs, vmax=vmax_abs)
    ax2.invert_yaxis()
    ax2.plot(np.ma.array(lats, mask=s95_r[exptest]), levs, 'k+', ms=2.5)
    ax2.set_xlabel('latitude')
    ax2.set_ylabel('height (hPa)')
    ax2.set_title(f'drmse RH ({exptest} - {expref}), Ref=IFS')
    ax2.set_ylim(1000,10)
    ax2.grid(True)
    # set colorbar
    cbar = fig.colorbar(contour2,location='bottom',aspect=40)
    cbar.set_label('drmse (%)')
    
    plt.savefig(f"IFS_drmse_RH_{exptest}.png")
```
