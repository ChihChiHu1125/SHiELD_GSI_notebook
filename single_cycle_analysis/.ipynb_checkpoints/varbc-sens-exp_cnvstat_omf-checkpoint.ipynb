```python
# plot omf/oma for cnvstat -- for channel sensitivity experiment
# initial built (copied from cnvstat_omf_single-cycle-varBC): 2024/07/17
# modified: 2024/07/31
# modified: 2025/02/05
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
instr  = ['t','q','gps','uv','amv']

cycle_tag = '_v2'  # '' or '_v2'
gesanl = 'anl'  # 'ges' or 'anl'

if cycle_tag == '_v2':
    if gesanl == 'ges':
        cycle_date = '2022073106'
    elif gesanl == 'anl':
        cycle_date = '2022073100'
else:
    if gesanl == 'ges':
        cycle_date = '2022070112'
    elif gesanl == 'anl':
        cycle_date = '2022070106'

sens_test = 'atms'

if sens_test == 'amsua':
    exps = ['amsua_ch1','amsua_ch2','amsua_ch3','amsua_ch4','amsua_ch5','amsua_ch15']
    datahome = f'/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/single_cycle_VarBC_sens_exp{cycle_tag}/obs_space_csv/cnv_CTRL-amsua_ch1-amsua_ch2-amsua_ch3-amsua_ch4-amsua_ch5-amsua_ch15'
elif sens_test == 'atms':
    exps = ['atms_ch17','atms_ch18','atms_ch19','atms_ch20','atms_ch21','atms_ch22']
    datahome = f'/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/single_cycle_VarBC_sens_exp{cycle_tag}/obs_space_csv/cnv_CTRL-atms_ch17-atms_ch18-atms_ch19-atms_ch20-atms_ch21-atms_ch22'
    

print(datahome)
print(exps)
print(gesanl)
print(cycle_date)
```

```python
df={}; # the stats

for i, instrument in enumerate(instr):
    if gesanl == 'anl':
        fn = f'{datahome}/cnvstats_{instrument}_{cycle_date}_anl.csv'
    elif gesanl == 'ges':
        fn = f'{datahome}/cnvstats_{instrument}_{cycle_date}_ges.csv'
    df[instrument] = pd.read_csv(fn)
```

```python
# loot at the original dataframe
#df['t']

# loot at the data counts
tmp  = df['uv']
tmpp = tmp[tmp['region']=='GLB']

columns_counts = tmpp.filter(like='counts_', axis=1).columns
tmpp[['pltlev'] + list(columns_counts)]

#tmpp[['pltlev','counts_ctrl','counts_amsua_ng_1','counts_amsua_ng_2']]
```

```python
# for analysis

save_fig = True
fig_dir = f'/home/Chih-Chi.Hu/python_codes/figures/single_cycle_VarBC_sens_exp/cnvstats/'

if not os.path.exists(fig_dir):
    os.makedirs(fig_dir)
    print(f"Directory '{fig_dir}' created.")
    
region     = 'GLB'


if gesanl == 'anl':
    for instrument in instr:
        df_tmp = df[instrument]

        stdnorm={};
        for exp in exps:
            stdnorm[exp] = df_tmp[df_tmp['region'] == region][f'rms_{exp}'].values

        pltlevs = df_tmp[df_tmp['region'] == region]['pltlev'].values
        pltlevs[pltlevs==1100] = 1030


        fig = plt.figure(figsize=(4.5,5))

        if instrument == 't':
            if sens_test == 'amsua': dx = 0.1
            if sens_test == 'atms':  dx = 0.05
            plt.ylim(1030,5)

        elif instrument == 'q':
            if sens_test == 'amsua': dx = 0.5
            if sens_test == 'atms':  dx = 0.1
            plt.ylim(1030,300)

        elif instrument == 'gps':
            if sens_test == 'amsua': dx = 0.2
            if sens_test == 'atms':  dx = 0.15
            plt.ylim(1000,0)

        elif instrument == 'uv':
            if sens_test == 'amsua': dx = 0.1
            if sens_test == 'atms':  dx = 0.05
            plt.ylim(1030,5)

        elif instrument == 'amv':
            if sens_test == 'amsua': dx = 0.1
            if sens_test == 'atms':  dx = 0.05
            plt.ylim(1000,100)

        for exp in exps:
            if instrument in ('gps','amv'):
                plt.plot(stdnorm[exp][:-1], pltlevs[:-1], linewidth=2.5, label=f'{exp}')
            else:
                plt.plot(stdnorm[exp], pltlevs, linewidth=2.5, label=f'{exp}')
        plt.xlim(100-dx, 100+dx)
        plt.axvline(100,color='black',linestyle='--')
        plt.grid(True)
        plt.title(f'{instrument} {region} ({gesanl}) [{cycle_date}]',fontsize=14)
        plt.xlabel('norm. O-A')
        plt.ylabel('(hPa)')
        plt.legend()
        plt.tight_layout()

        if save_fig:
            plt.savefig(f"{fig_dir}/{instrument}_{gesanl}_{region}_{sens_test}_{cycle_date}.png")
        plt.show()
```
