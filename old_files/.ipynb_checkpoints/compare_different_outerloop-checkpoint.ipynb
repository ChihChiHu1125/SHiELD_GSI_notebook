```python
# check if the symmetric cloud 'evolves' in each outer loop:

import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
```

```python
# the first outer loop
rr_ges=nc.Dataset(f"/home/Chih-Chi.Hu/python_codes/diag_amsua_n15_ges.2022073000.nc4")

# the analysis (after the second outer loop)
rr_anl=nc.Dataset(f"/home/Chih-Chi.Hu/python_codes/diag_amsua_n15_anl.2022073000.nc4")
```

```python
ch_ges         = rr_ges.variables['Channel_Index'][:]
inv_obserr_ges = rr_ges.variables['Inverse_Observation_Error'][:]
obserr_ges     = rr_ges.variables['Sym_Observation_Error'][:]
clw_obs_ges    = rr_ges.variables['clw_obs'][:]
clw_model_ges  = rr_ges.variables['clw_guess_retrieval'][:]
omf_ges        = rr_ges.variables['Obs_Minus_Forecast_adjusted'][:]
wspd_ges       = rr_ges.variables['Sfc_Wind_Speed'][:]
fland_ges      = rr_ges.variables['Land_Fraction'][:]
fwater_ges     = rr_ges.variables['Water_Fraction'][:]
symcloud_ges   = 0.5*(clw_obs_ges + clw_model_ges)
qc_ges         = rr_ges.variables['QC_Flag'][:]

ch_anl         = rr_anl.variables['Channel_Index'][:]
inv_obserr_anl = rr_anl.variables['Inverse_Observation_Error'][:]
obserr_anl     = rr_anl.variables['Sym_Observation_Error'][:]
clw_obs_anl    = rr_anl.variables['clw_obs'][:]
clw_model_anl  = rr_anl.variables['clw_guess_retrieval'][:]
omf_anl        = rr_anl.variables['Obs_Minus_Forecast_adjusted'][:]
wspd_anl       = rr_anl.variables['Sfc_Wind_Speed'][:]
fland_anl      = rr_anl.variables['Land_Fraction'][:]
fwater_anl     = rr_anl.variables['Water_Fraction'][:]
symcloud_anl   = 0.5*(clw_obs_anl + clw_model_anl)
qc_anl         = rr_anl.variables['QC_Flag'][:]
```

```python
chn=1

sea_screen_ges  = (fwater_ges>= 0.99)
land_screen_ges = (fland_ges >= 0.01)
use_ges = np.where((qc_ges == 0)&(ch_ges==chn)&sea_screen_ges)

sea_screen_anl  = (fwater_anl>= 0.99)
land_screen_anl = (fland_anl >= 0.01)
use_anl = np.where((qc_anl == 0)&(ch_anl==chn)&sea_screen_anl)

# take the union of the two
use= np.intersect1d(use_ges, use_anl)

#print(use[0].shape)
obserr_use_ges     = obserr_ges[use]
inv_obserr_use_ges = inv_obserr_ges[use]
cloud_use_ges      = symcloud_ges[use]
omf_use_ges        = omf_ges[use]
wspd_use_ges       = wspd_ges[use]
clw_obs_use_ges    = clw_obs_ges[use] # for sanity check

obserr_use_anl     = obserr_anl[use]
inv_obserr_use_anl = inv_obserr_anl[use]
cloud_use_anl      = symcloud_anl[use]
omf_use_anl        = omf_anl[use]
wspd_use_anl       = wspd_anl[use]
clw_obs_use_anl    = clw_obs_anl[use] # for sanity check
```

```python
# sanity check -- to see if the obs clouds are exactly the same between different outer loops:
plt.scatter(clw_obs_use_ges, clw_obs_use_anl)
plt.plot([0,1],[0,1])
plt.grid(True)
```

```python
# check if the symmetric cloud change
plt.scatter(cloud_use_ges, cloud_use_anl)
plt.xlabel('1st outer loop symmetric cloud')
plt.ylabel('analysis symmetric cloud')
plt.plot([0,1],[0,1])
plt.grid(True)
```

```python
# check if the errors are changed
plt.scatter(obserr_use_ges, obserr_use_anl)
plt.xlabel('1st outer loop original error')
plt.ylabel('analysis original error')
plt.plot([0,10],[0,10])
plt.grid(True)
```

```python
# see the difference between the final errors
plt.scatter(1/inv_obserr_use_ges, 1/inv_obserr_use_anl)
plt.xlabel('1st outer loop final error')
plt.ylabel('analysis final error')
plt.plot([0,48],[0,48])
plt.grid(True)
```

```python
# check if both of the orignal errors follow the piecewise linear fit
plt.scatter(cloud_use_ges, obserr_use_ges)
plt.scatter(cloud_use_anl, obserr_use_anl)
plt.xlabel('symmetric cloud')
plt.ylabel('stdev')
plt.grid(True)
```
