```python
# Plot the verification against reanalysis data (for IFS only):
# Plot normalized 'drmse' in lat-height coordinate
# initial built: 2024/09/23
# modified: add U,V: 2024/10/10
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import netCDF4 as nc
```

```python
datahome    = '/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/GribDiag/ush/plotgribnc/data/gribnc'

timerange = '2022061500_2022072000_12'
expname   = 'ctrl-ctrl_noinf'
exps = expname.split('-')
expref    = exps[0]


ncdir = f'{datahome}/IFS/{timerange}/{expname}'
```

```python
# read netcdfs:
rmse_r = {}; rmse_t={}; rmse_u={}; rmse_v={};
bias_r = {}; bias_t={}; bias_u={}; bias_v={};

for exp in exps:
    if exp != expref:
        # normalized relative error:
        f = nc.Dataset(f"{ncdir}/dzmean_{exp}_f06.nc")
        lat = f.variables['latitude'][:]
        lev = f.variables['level'][:]
        dr_norm[exp] = f.variables['r'][:]
        dt_norm[exp] = f.variables['t'][:]
        du_norm[exp] = f.variables['u'][:]
        dv_norm[exp] = f.variables['v'][:]

    fn_bias = nc.Dataset(f"{ncdir}/bias_{exp}_f06.nc")
    bias_r[exp] = fn_bias.variables['r'][:]
    bias_t[exp] = fn_bias.variables['t'][:]
    bias_u[exp] = fn_bias.variables['u'][:]
    bias_v[exp] = fn_bias.variables['v'][:]
```
