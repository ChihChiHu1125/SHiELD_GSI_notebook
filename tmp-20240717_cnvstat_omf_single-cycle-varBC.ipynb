```python
# plot omf/oma for cnvstat
# initial built: 2024/07/17
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
instr  = ['t','q','gps','uv','amv']
exps   = ['bc0_noinf', 'bc0_ng', 'bc1_inf', 'bc1_noinf', 'bc1_ng']
#fn_dir = '/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/single_cycle_VarBC_experiment/obs_space_csv/bc0_inf-bc0_noinf-bc1_inf-bc1_noinf'
#fn_dir = '/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/old_crtm_VarBC_experiment/obs_space_csv/bc0_inf-bc0_noinf-bc1_inf-bc1_noinf'
fn_dir = '/scratch2/GFDL/gfdlscr/Chih-Chi.Hu/single_cycle_VarBC_experiment/obs_space_csv/bc0_inf-bc0_noinf-allobs-bc0_ng-allobs-bc1_inf-allobs-bc1_noinf-allobs-bc1_ng-allobs'
```

```python
df={}; # the stats

for i, instrument in enumerate(instr):
    fn = f'{fn_dir}/cnvstats_{instrument}_anl.csv'
    df[instrument] = pd.read_csv(fn)
```

```python
# loot at the original dataframe
#df['t']

# loot at the data counts
tmp  = df['t']
tmpp = tmp[tmp['region']=='GLB']

tmpp[['pltlev','counts_bc0_inf']]
```

```python
instrument = 'gps'
region     = 'GLB'
df_tmp = df[instrument]

stdnorm={};
for exp in exps:
    stdnorm[exp] = df_tmp[df_tmp['region'] == region][f'rms_{exp}'].values
pltlevs = df_tmp[df_tmp['region'] == region]['pltlev'].values
pltlevs[pltlevs==1100] = 1030
```

```python
region     = 'GLB'

for instrument in instr:
    df_tmp = df[instrument]

    stdnorm={};
    for exp in exps:
        stdnorm[exp] = df_tmp[df_tmp['region'] == region][f'rms_{exp}'].values
        
    pltlevs = df_tmp[df_tmp['region'] == region]['pltlev'].values
    pltlevs[pltlevs==1100] = 1030
    
    
    fig = plt.figure(figsize=(4.5,5))

    if instrument == 't':
        dx = 0.3
        plt.ylim(1030,5)
        
    elif instrument == 'q':
        dx = 3
        plt.ylim(1030,300)
        
    elif instrument == 'gps':
        dx = 3.5
        plt.ylim(1000,0)
        
    elif instrument == 'uv':
        dx = 0.8
        plt.ylim(1030,5)

    elif instrument == 'amv':
        dx = 2.5
        plt.ylim(1000,100)

    for exp in exps:
        if instrument in ('gps','amv'):
            plt.plot(stdnorm[exp][:-1], pltlevs[:-1], linewidth=2.5, label=f'{exp}')
        else:
            plt.plot(stdnorm[exp], pltlevs, linewidth=2.5, label=f'{exp}')
    plt.xlim(100-dx, 100+dx)
    plt.axvline(100,color='black',linestyle='--')
    plt.grid(True)
    plt.title(f'{instrument} {region}',fontsize=14)
    plt.xlabel('norm. O-A')
    plt.ylabel('(hPa)')
    plt.legend()
    
    plt.savefig(f"cnv_omarms_{instrument}-{region}.png")
```
